```python
from PIL import Image
import os
```

```python

def resize_images(input_folder, output_folder, new_size=(256, 256)):
    # 입력 폴더가 존재하는지 확인하고 없다면 생성
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 입력 폴더 내의 모든 파일 목록을 가져옴
    image_files = [f for f in os.listdir(input_folder) if f.lower().endswith((".jpg", ".jpeg", ".png", ".gif", ".bmp"))]

    for image_file in image_files:
        try:
            # 이미지를 열고 크기 조정
            img = Image.open(os.path.join(input_folder, image_file))
            img = img.resize(new_size, Image.ANTIALIAS)  # ANTIALIAS를 사용하여 이미지 품질 향상

            # 변경된 이미지를 저장
            img.save(os.path.join(output_folder, image_file))
            print(f"{image_file}의 크기가 변경되었습니다.")
        except Exception as e:
            print(f"오류: {image_file}를 처리하는 중에 오류가 발생했습니다. ({e})")

    print("작업 완료")

```

```python
# 사용 예시
input_folder = "./상추__균핵병"
output_folder = "./new/"
resize_images(input_folder, output_folder)
```
